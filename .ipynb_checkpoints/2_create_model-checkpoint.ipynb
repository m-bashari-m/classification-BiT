```python
import os
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
BASE_DIR = '/content/drive/MyDrive/data'
train_dir = os.path.join(BASE_DIR, 'train')
val_dir = os.path.join(BASE_DIR, 'val')

classes = os.listdir(train_dir)
n_classes = len(classes)
print(classes)
```

# Prepare data

```python
IMAGE_SIZE = (128,128)
BATCH_SIZE = 64
AUTOTUNE = tf.data.AUTOTUNE
BUFFER_SIZE = 1000
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(train_dir,
                                                      shuffle=True,
                                                      label_mode='categorical',
                                                      batch_size=BATCH_SIZE,
                                                      image_size=IMAGE_SIZE)
val_ds = tf.keras.utils.image_dataset_from_directory(val_dir,
                                                      label_mode='categorical',
                                                      batch_size=BATCH_SIZE,
                                                      image_size=IMAGE_SIZE)
```

```python
for batch, lbl in train_ds.take(1):
    print(batch.shape, lbl.shape)
    plt.axis('off')
    plt.imshow(batch[0]/255.0)
```

```python
CROP_SIZE = 128
RESIZE_TO = (135, 135)
```

```python
train_ds = train_ds.repeat(2).prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)
```

# Create model

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomTranslation((-.1, .1),
                                      (-.1, .1),
                                      fill_mode='constant'),
    tf.keras.layers.RandomCrop(IMAGE_SIZE[0], IMAGE_SIZE[1]),
    tf.keras.layers.RandomContrast(.3) 
])
```

```python
for image, _ in train_ds.take(1):
  plt.figure(figsize=(5, 5))
  first_image = image[0]
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    img = data_augmentation(tf.expand_dims(first_image, 0))
    img = img[0] / 255.
    plt.imshow(img)
    plt.axis('off')
```

```python
base_model = tf.keras.applications.efficientnet_v2.EfficientNetV2B0(
    include_top=False, input_shape=(IMAGE_SIZE + (3,))
)

base_model.trainable = False
```

```python
input = tf.keras.layers.Input(shape=IMAGE_SIZE + (3,))
x = tf.keras.applications.efficientnet_v2.preprocess_input(input)
x = data_augmentation(x)
x = base_model(x, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
output = tf.keras.layers.Dense(3)(x)

model = tf.keras.Model(inputs=[input], outputs=[output])
```

```python
optimizer = tf.keras.optimizers.Adam()
loss_fn = tf.keras.losses.CategoricalCrossentropy(from_logits=True)

model.compile(optimizer=optimizer,
             loss=loss_fn,
             metrics=['accuracy'])
```

# Train model

```python
!mkdir checkpoint
```

```python
CHECKPOINT_DIR = os.path.join('checkpoint')
```

```python
fine_tuning_callbacks = [
    tf.keras.callbacks.ModelCheckpoint(CHECKPOINT_DIR, save_best_only=True),
    tf.keras.callbacks.EarlyStopping(patience=3)                       
]
```

```python
model.summary()
```

```python
history = model.fit(train_ds,
                   validation_data=val_ds,
                   epochs=10)
```

```python
model.save('b-fine-tune.h5')
```
